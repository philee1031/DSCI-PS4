```python
import tensorflow as tf
import numpy as np
import pandas as pd
import random as rd
import sklearn.datasets
import sklearn.cluster
import sklearn.decomposition
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
from numpy import expand_dims
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import RMSprop
import os
import PIL
```

# Load Data

```python
df = np.load('ps4_trainvalid_images.npy')
labels = pd.read_csv("ps4_trainvalid_labels.csv")['Label'].to_numpy()
```

```python
df = df/255.0
df.shape, labels.shape
```

```python
# Create index

idx = list(range(len(labels)))
np.random.seed(1)
np.random.shuffle(idx)

# Split data using the index created (8:2)

N = len(idx)
train_idx = idx[:int(N*0.8)]
valid_idx = idx[int(N*0.8):]

train_df = df[train_idx]
valid_df = df[valid_idx]

train_labels = labels[train_idx]
valid_labels = labels[valid_idx]

train_df.shape, valid_df.shape, train_labels.shape, valid_labels.shape
```

```python
# Case distribution

x = labels[:]
y = train_labels[:]
import collections
orig=collections.Counter(x)
train=collections.Counter(y)
print(orig, train)
```

** It is expected to do some data augmentation. The share of interested case (plueral effusion) in the given data is 2.88 percent. The share of preexisting condition in the given data is 17.89 percent. Hence, without data augmentation, the model may incorrectly identify two of three cases.

```python
# One-hot Encoding

train_labelshot = tf.one_hot(train_labels, 3).numpy()
valid_labelshot = tf.one_hot(valid_labels, 3).numpy()

train_labelshot.shape, valid_labelshot.shape
```

# Data Augmentation

```python
# Create Image Data Generator for Train Set

data_augmentation = Sequential([
    layers.experimental.preprocessing.RandomFlip("horizontal"),
    layers.experimental.preprocessing.RandomContrast(0.02),
    layers.experimental.preprocessing.RandomTranslation(0.05, 0.05),
    layers.experimental.preprocessing.RandomRotation(0.02),
    layers.experimental.preprocessing.RandomZoom((-0.02, 0.02), fill_mode='reflect'),
])
```

```python
# Separate Data into condition 1 and 2

mask1=train_labels == 1
mask2=train_labels ==2

predf, prelabels = train_df[mask1], train_labels[mask1]
effdf, efflabels = train_df[mask2], train_labels[mask2]

bal_train_df = train_df.copy()
bal_train_labels = train_labels.copy()

# Create data augmented scans

for _ in range(3):
    bal_train_df = np.vstack([bal_train_df, data_augmentation(predf)])
    bal_train_labels = np.hstack([bal_train_labels, prelabels])

for _ in range(25):
    bal_train_df = np.vstack([bal_train_df, data_augmentation(effdf)])
    bal_train_labels = np.hstack([bal_train_labels, efflabels])
    
collections.Counter(bal_train_labels)
```

```python
# One-hot Encoding

bal_train_labelshot = tf.one_hot(bal_train_labels, 3).numpy()

bal_train_labelshot.shape
```

# CNN Models

```python
# Attempt 1

model1 = Sequential([
    layers.Conv2D(
        filters=16, kernel_size=(3,3), strides=1, activation='relu', input_shape=(64,64,1)),
    layers.MaxPool2D(pool_size=(2,2), strides=None, padding='valid'),
    layers.Conv2D(
        filters=32, kernel_size=(3,3), strides=1, activation='relu'),
    layers.MaxPool2D(pool_size=(2,2), strides=None, padding='valid'),
    layers.Conv2D(
        filters=64, kernel_size=(3,3), strides=1, activation='relu'),
    layers.MaxPool2D(pool_size=(2,2), strides=None, padding='valid'),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(3, activation='softmax')
])

model1.summary()
```

```python
model1.compile(
    optimizer = 'adam',
loss=tf.keras.losses.CategoricalCrossentropy(from_logits=False),
metrics=['accuracy', 'AUC'])
```

```python
# Attempt 2

model2 = Sequential([
    layers.Conv2D(
        filters=32, kernel_size=(3,3), strides=1, activation='relu', input_shape=(64,64,1)),
    layers.MaxPool2D(pool_size=(2,2), strides=None, padding='valid'),
    layers.Conv2D(
        filters=32, kernel_size=(3,3), strides=1, activation='relu'),
    layers.MaxPool2D(pool_size=(2,2), strides=None, padding='valid'),
    layers.Conv2D(
        filters=64, kernel_size=(3,3), strides=1, activation='relu'),
    layers.MaxPool2D(pool_size=(2,2), strides=None, padding='valid'),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(3, activation='softmax')
])

model2.summary()
```

```python
model2.compile(
    optimizer = 'adam',
loss=tf.keras.losses.CategoricalCrossentropy(from_logits=False),
metrics=['accuracy', 'AUC'])
```

```python
# Attempt 4

model4 = Sequential([
    layers.Conv2D(
        filters=32, kernel_size=(3,3), strides=1, activation='relu', input_shape=(64,64,1)),
    layers.MaxPool2D(pool_size=(2,2), strides=None, padding='valid'),
    layers.Conv2D(
        filters=32, kernel_size=(3,3), strides=1, activation='relu'),
    layers.MaxPool2D(pool_size=(2,2), strides=None, padding='valid'),
    layers.Conv2D(
        filters=64, kernel_size=(3,3), strides=1, activation='relu'),
    layers.MaxPool2D(pool_size=(2,2), strides=None, padding='valid'),
    layers.Conv2D(
        filters=64, kernel_size=(3,3), strides=1, activation='relu'),
    layers.MaxPool2D(pool_size=(2,2), strides=None, padding='valid'),    
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),    
    layers.Dense(3, activation='softmax')
])

model4.summary()
```

```python
model4.compile(
    optimizer = 'adam',
loss=tf.keras.losses.CategoricalCrossentropy(from_logits=False),
metrics=['accuracy', 'AUC'])
```
